```python
fiscal_encoded = pd.get_dummies(apple_df['Fiscal Period'], dtype = int)
fiscal_encoded
```

```python
#Drop the fiscal period column and concatinate your encoded column to the original df
apple_df = apple_df.drop(columns = ['Fiscal Period'])
#Concatinate
apple_df = pd.concat([apple_df, fiscal_encoded], axis = 1)
apple_df
```

```python
#Feature engineering and next quarter EPS calculation
apple_df.info()
```

```python
#Sort in ascending order using the report date
apple_df = apple_df.sort_values(by = "Report Date").reset_index(drop = True)
apple_df
```

```python
#Engineer a % change feature, capturing change in revenue from one quarter to the next
apple_df['Revenue' + 'Change From Last Quarter']= apple_df['Revenue'].pct_change()
apple_df
```

```python
#Engineer other features to capture different measures in bulk
features_to_get_delta = ['Shares (Basic)',
                        'Shares (Diluted)',
                        'Revenue',
                        'Operating Income (Loss)',
                        'Net Income',
                        'Total Liabilities & Equity']
#pct_change() calculated the change from last quarter to the current
for feature in features_to_get_delta:
    apple_df[feature + 'change from last quarter'] = apple_df[feature].pct_change()

apple_df
```

```python
#Create a row for the next quarter EPS 
apple_df['Next Quarter EPS']= apple_df['Earnings Per Share, Basic'].shift(-1)
apple_df
```

```python
#Drop null elements in the engineered features, first next quarter EPS
apple_df = apple_df.dropna(subset = ['Next Quarter EPS']).reset_index(drop = True)
apple_df
```

```python
#Drop for the first row
apple_df = apple_df.dropna().reset_index(drop = True)
apple_df
```

```python
#the target label for the ML regression model is the quarterly % change in EPS
#create a function that calculated the % change in the next quarters EPS
def create_regression_label(next_quarter, current_quarter):
    #if current EPS is 0 we set the label to 0
    if abs (current_quarter) ==0:
       return 0
    return(next_quarter - current_quarter)/ abs (current_quarter)
apple_df['% Quarterly change in EPS(Regression Label)'] =\
apple_df.apply(lambda row: create_regression_label(row['Next Quarter EPS'], row['Earnings Per Share, Basic']), axis = 1)
apple_df
```

```python
#Drop null elements in the engineered features, first next quarter Revenue
apple_df = apple_df.dropna(subset = ['Revenue']).reset_index(drop = True)
apple_df
```
